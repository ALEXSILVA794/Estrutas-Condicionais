```python
# Contador de 1 a 100 usando FOR (somente pares)

print("Números pares de 1 a 100 usando FOR:")
for i in range(1, 101):
    if i % 2 == 0:  # verifica se é par
        print(i)

# Contador de 1 a 100 usando WHILE (somente pares)

print("Números pares de 1 a 100 usando WHILE:")
num = 1

while num <= 100:
    if num % 2 == 0:  # verifica se é par
        print(num)
    num += 1
```
