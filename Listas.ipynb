```python
# Programa que recebe nomes de alunos e armazena em uma lista

alunos = []  # lista vazia

print("Digite os nomes dos alunos. Para encerrar, digite 'sair'.")

while True:
    nome = input("Nome do aluno: ")

    if nome.lower() == "sair":
        break

    alunos.append(nome)  # adiciona nome Ã  lista

print("\nLista de alunos cadastrados:")
for aluno in alunos:
    print(f"- {aluno}")





```
